```python
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()

# Print the numerical columns
print("Numerical columns:", numerical_columns)
```

```python
import math
mean_value=df[numerical_columns].mean()
mode_value=df[numerical_columns].mode().iloc[0]
median_value=df[numerical_columns].median()
standard_deviation=df[numerical_columns].std()
print( "mean is :\n" ,mean_value)
print( "mode is:\n" ,mode_value)
print( "median is :\n" ,median_value)
print( "sd is:\n",standard_deviation)
```

Mean indicating the average sales over a specific period. This helps in understanding the general trend and making comparisons.the mean can be used to create new features, such as the difference between a particular sales value and the average sales, which might help in identifying outliers or trends.

 Understanding the median helps in assessing the skewness of the data. If the mean and median are significantly different, it indicates a skewed distribution

 The standard deviation quantifies the amount of variation or dispersion in the sales data. It helps in understanding how spread out the sales values are around the mean.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    # Analyzing the distribution
    skewness = df[column].skew()
    outliers = df[column][(df[column] < (df[column].quantile(0.25) - 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)))) |
                         (df[column] > (df[column].quantile(0.75) + 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25))))]

    print(f'Analysis for {column}:')
    print(f'Skewness: {skewness}')
    print(f'Outliers: {outliers.tolist()}\n')
```

```python
  for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(df[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show()
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)]

    print(f'Analysis for {column}:')
    print(f'IQR: {IQR}')
    print(f'Lower Bound: {lower_bound}')
    print(f'Upper Bound: {upper_bound}')
    print(f'Outliers: {outliers.tolist()}\n')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the DataFrame only contains numeric values and handle NaNs
df = df.apply(pd.to_numeric, errors='coerce')



# Plot before standardization
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['Total Sales Value'], kde=True)
plt.title('Total Sales Value - Before Standardization')

plt.subplot(1, 3, 2)
sns.histplot(df['Net Sales Value'], kde=True)
plt.title('Net Sales Value - Before Standardization')

plt.subplot(1, 3, 3)
sns.histplot(df['Discount Rate (%)'], kde=True)
plt.title('Discount Rate (%) - Before Standardization')

plt.tight_layout()
plt.show()

# Standardization
df_standardized = df.copy()

for column in df.columns:
    mean = df[column].mean()
    std_dev = df[column].std()
    df_standardized[column] = (df[column] - mean) / std_dev

# Plot after standardization
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.histplot(df_standardized['Total Sales Value'], kde=True)
plt.title('Total Sales Value - After Standardization')

plt.subplot(1, 3, 2)
sns.histplot(df_standardized['Net Sales Value'], kde=True)
plt.title('Net Sales Value - After Standardization')

plt.subplot(1, 3, 3)
sns.histplot(df_standardized['Discount Rate (%)'], kde=True)
plt.title('Discount Rate (%) - After Standardization')

plt.tight_layout()
plt.show()

df_standardized
```
