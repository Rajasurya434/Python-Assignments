```python
import pandas as pd
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score



# Model 1: Using feature1 and feature2
X1_train = train_data[['Age_08_04', 'KM']]
y1_train = train_data['HP']
X1_test = test_data[['Age_08_04', 'KM']]
y1_test = test_data['HP']
X1_train = sm.add_constant(X1_train)
X1_test = sm.add_constant(X1_test)
model1 = sm.OLS(y1_train, X1_train).fit()

# Model 2: Using feature1 and feature3
X2_train = train_data[['Age_08_04', 'cc']]
y2_train = train_data['HP']
X2_test = test_data[['Age_08_04', 'cc']]
y2_test = test_data['HP']
X2_train = sm.add_constant(X2_train)
X2_test = sm.add_constant(X2_test)
model2 = sm.OLS(y2_train, X2_train).fit()

# Model 3: Using feature2 and feature3
X3_train = train_data[['KM', 'cc']]
y3_train = train_data['HP']
X3_test = test_data[['KM', 'cc']]
y3_test = test_data['HP']
X3_train = sm.add_constant(X3_train)
X3_test = sm.add_constant(X3_test)
model3 = sm.OLS(y3_train, X3_train).fit()

# Making predictions on the test set
y1_pred = model1.predict(X1_test)
y2_pred = model2.predict(X2_test)
y3_pred = model3.predict(X3_test)

# Evaluating the models
def evaluate_model(y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mae, mse, r2

mae1, mse1, r21 = evaluate_model(y1_test, y1_pred)
mae2, mse2, r22 = evaluate_model(y2_test, y2_pred)
mae3, mse3, r23 = evaluate_model(y3_test, y3_pred)

print(f"Model 1 - MAE: {mae1}, MSE: {mse1}, R2: {r21}")
print(f"Model 2 - MAE: {mae2}, MSE: {mse2}, R2: {r22}")
print(f"Model 3 - MAE: {mae3}, MSE: {mse3}, R2: {r23}")
```

Comparing between above three models the third model performs well because MAE and MSE values are lower for third model compared to other two models and Rsquare value is highier indicating better fit of data to the model .

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X_train = train_data[['Age_08_04', 'KM', 'cc']]
y_train = train_data['HP']
X_test = test_data[['Age_08_04', 'KM', 'cc']]
y_test = test_data['HP']

#  Lasso regression
lasso = Lasso(alpha=1.0)
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)

#  Ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

def evaluate_model(y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mae, mse, r2

mae_lasso, mse_lasso, r2_lasso = evaluate_model(y_test, y_pred_lasso)
mae_ridge, mse_ridge, r2_ridge = evaluate_model(y_test, y_pred_ridge)

print(f"Lasso Regression - MAE: {mae_lasso}, MSE: {mse_lasso}, R2: {r2_lasso}")
print(f"Ridge Regression - MAE: {mae_ridge}, MSE: {mse_ridge}, R2: {r2_ridge}")
```

 This suggests that, on average, Lasso regression provides more accurate predictions with smaller errors. It means Lasso might be better at reducing the magnitude of prediction errors on average.

This indicates that Ridge regression reduces larger errors more effectively than Lasso. Since MSE squares the errors before averaging them, it penalizes larger errors more heavily

This suggests that Ridge regression explains a slightly larger proportion of the variance in the dependent variable. A higher R² value indicates a better overall fit of the model to the data.
