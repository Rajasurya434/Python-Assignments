```python
df=df.drop_duplicates()
```

```python
df=df.dropna()
```

```python

df.columns = ['Transaction']
```

```python
transactions = df['Transaction'].apply(lambda x: x.split(','))

transactions = transactions.apply(lambda x: [item.strip() for item in x])

print(transactions.head())
```

Splits the string into list of items taking ',' to seperate

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()

te_ary = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_ary, columns=te.columns_)

basket.head()
```

Convert to binary matrix so that it can be used with aprori algorthmn , we converted raw data to binary matrix .

```python
from mlxtend.frequent_patterns import apriori, association_rules


frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)


rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)


rules.head()
```

```python
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

rules = rules[rules['lift'] > 1.0]
print(rules.head())
```

This fliters data with confidence >0.5 and lift > 1
